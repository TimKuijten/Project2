```python
#We have 125000 rows from lots of cities in Europe, as we only need Barcelona we drop all rows that arent Barcelona in City
valid_city = ['Barcelona']
df.loc[~df['City'].isin(valid_city), 'City'] = np.nan

df = df.dropna(subset=['City'])
# Now we have 8425 restaurants left, only in Barcelona

# Dropping all columns we don't need
df.drop(columns='Ranking',inplace=True)
df.drop(columns='Reviews',inplace=True)
df.drop(columns='ID_TA',inplace=True)
df.drop(columns='URL_TA',inplace=True)
df.drop(columns='Price Range',inplace=True)


df
```
