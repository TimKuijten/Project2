```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df_glovo = df_glovo.replace('New', np.nan).replace('--', np.nan)
df_glovo['Rating Glovo'] = df_glovo['Rating Glovo'].astype(str)
df_glovo['Rating Glovo'] = pd.to_numeric(df_glovo['Rating Glovo'].str.strip('%'), errors='coerce')

# Split cuisine styles into individual values and create a new data frame
cuisine_styles = df_glovo['Cuisine Style'].apply(lambda x: x.replace("'", '').replace('[','').replace(']','').replace(' ','').split(','))
cuisine_df = pd.DataFrame(cuisine_styles.tolist(), columns=['Cuisine Style 1', 'Cuisine Style 2'])

# Explode the new data frame so that each row is a single cuisine style and not a list
cuisine_df = cuisine_df['Cuisine Style'].apply(pd.Series).merge(df_glovo, right_index = True, left_index = True).drop(["Cuisine Style"], axis = 1).melt(id_vars = ['Name', 'Rating Glovo', 'Number of Reviews'], value_name = "Cuisine Style").drop("variable", axis = 1).dropna()

# Plot the average rating by cuisine style, only include the top 15 styles by frequency
top_15_cuisines = cuisine_df['Cuisine Style'].value_counts().nlargest(15).index.tolist()
cuisine_df = cuisine_df[cuisine_df['Cuisine Style'].isin(top_15_cuisines)]

# Calculate the average rating for each cuisine style and plot it using seaborn
sns.barplot(x='Cuisine Style', y='Rating Glovo', data=df_glovo.merge(cuisine_df, on='Cuisine Style'))
plt.xticks(rotation=90)
plt.show()
```
