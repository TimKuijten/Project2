```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_glovo_ta = pd.read_csv(r'../../data/df_glovo_ta.csv')
df_glovo = pd.read_csv(r'../../data/df_glovo.csv')
df_ta = pd.read_csv(r'../../data/df_ta.csv')
```

```python
df_glovo_ta
```

```python
df_glovo_ta['Rating Glovo'] = df_glovo_ta['Rating Glovo'].replace("--", np.nan)
df_glovo_ta['Rating Glovo'] = df_glovo_ta['Rating Glovo'].replace("New", np.nan)
df_glovo_ta['Rating Glovo'] = df_glovo_ta['Rating Glovo'].str.strip('%').astype(float)
df_glovo_ta['Rating Glovo'] = df_glovo_ta['Rating Glovo'].astype(float)
```

```python
df_glovo_ta['Rating TA'] = df_glovo_ta['Rating TA'].replace("--", np.nan)
df_glovo_ta['Rating TA'] = df_glovo_ta['Rating TA'].replace("New", np.nan)
df_glovo_ta['Rating TA'] = df_glovo_ta['Rating TA'].str.strip('%').astype(float)
df_glovo_ta['Rating TA'] = df_glovo_ta['Rating TA'].astype(float)
```

```python
df_glovo_ta['Rating TA'] = df_glovo_ta['Rating TA'].replace(-20, np.nan)
df_glovo_ta
```

```python
# Calculate the mean of 'Rating TA' and 'Rating Glovo' ignoring NaN values
mean_rating_ta = df_glovo_ta['Rating TA'].mean()
mean_rating_glovo = df_glovo_ta['Rating Glovo'].mean()
mean_rating_ta = round(mean_rating_ta)
mean_rating_glovo = round(mean_rating_glovo)
# Create a bar plot with the means
fig, ax = plt.subplots(figsize=(10,6))
sns.barplot(x=['Rating TA', 'Rating Glovo'], y=[mean_rating_ta, mean_rating_glovo], palette=["#00af87", "#ffc244"], ax=ax)
ax.set_title('Mean of rating of restaurants that are on both platforms')
ax.bar_label(ax.containers[0])
plt.show()
```
